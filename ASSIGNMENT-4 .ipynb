5.Splitting Dependent and Independent variables

```python
x = df.drop(columns=['Attrition'],axis=1)
x.head()
```

```python
y=df.Attrition
y.head()
```

6.Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x["BusinessTravel"]=le.fit_transform(x["BusinessTravel"])
```

```python
x["Department"]=le.fit_transform(x["Department"])
```

```python
x["Gender"]=le.fit_transform(x["Gender"])
```

```python
x["JobRole"]=le.fit_transform(x["JobRole"])
```

```python
x["MaritalStatus"]=le.fit_transform(x["MaritalStatus"])
```

```python
x["EducationField"]=le.fit_transform(x["EducationField"])
```

```python
x["Over18"]=le.fit_transform(x["Over18"])
```

```python
x["OverTime"]=le.fit_transform(x["OverTime"])
```

```python
x.info()
```

7.Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_scaled=pd.DataFrame(ms.fit_transform(x),columns=x.columns)
```

```python
x_scaled.head()
```

8.SPLITTING OF TRAINING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# MODEL BUILDING using decision tree


```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
pred=dtc.predict(x_test)
pred
```

```python
y_test
```

```python
x.iloc[:,0:16]
```

```python
y.head()
```

# Evaluation of classification model

```python
#Accuracy score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
pd.crosstab(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

# Roc-AUC curve

```python
probability=dtc.predict_proba(x_test)[:,1]
```

```python
probability
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, probability, pos_label='Yes')
```

```python
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```
